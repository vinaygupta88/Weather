```python
# import essential Libraries
import numpy as np
import pandas as pd
```

```python
#Acquire DataSet
df=pd.read_csv("dataset.csv")
df
```

```python
df.shape
```

```python
df2=df.drop("date",axis=1)
df2.head()
```

```python
df2.shape #number of row-1461 and number of column=5
```

# separating dataset into independent and dependent variable by using iloc functions

```python
X=df2.iloc[:,:4].values
X
```

```python
Y=df2.iloc[:,4].values
Y
```

# Spliting dataset into training and testing dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

# Applying Logistic Algorithm 

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

# Testing the Model

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
x_test
```

# Building confusion matrix 
to compare real and predicted values

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
